## Players who scored Test 100s most frequently
Qualification: Top 75 players ordered by Test 100s

### Load [espncricinfo's most test 100s](http://stats.espncricinfo.com/CI/content/records/227046.html) into a pandas dataframe

```python
import pandas as pd

df = pd.read_html("http://stats.espncricinfo.com/CI/content/records/227046.html")[0]
```

### Obviously Tendulkar is on top!

```python
df
```

### Caldulate innings per hundred
Substract not outs from total innings (similar to calculating batting average)

```python
df["inns_per_hundred"] = (df.Inns - df.NO) / df["100"]
```

### Obvoiusly Bradman is on top now!

```python
df.sort_values("inns_per_hundred")
```

### Draw the scatterplot
X-Axis: Number of 100s,
Y-Axis: Innings per 100

```python
%matplotlib inline

import matplotlib.pyplot as plt

df.player_surname = df.Player.str.split(" ").str[1]

hundreds = df["100"]
inns_per_hundred = df.inns_per_hundred
player = df.player_surname

plt.figure(figsize=(25, 25))
plt.scatter(hundreds, inns_per_hundred)

for label, x, y in zip(player, hundreds, inns_per_hundred):
    plt.annotate(label, xy = (x + 0.2, y), fontsize = 10)
```
